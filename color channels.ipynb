```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("Flowers.jpg",1)
color = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Splitting images into BGR channels using numpy

blue = color[:,:,2]
green = color[:,:,1]
red = color[:,:,0]
```

```python
plt.imshow(color)
plt.show()
```

```python
plt.imshow(blue)
plt.show()
```

```python
plt.imshow(green)
plt.show()
```

```python
plt.imshow(red)
plt.show()
```

```python
bgrb = cv2.merge((blue,green,red,blue))
bgrg = cv2.merge((blue,green,red,green))
bgrr = cv2.merge((blue,green,red,red))
```

```python
plt.imshow(bgrb)
plt.show()
```

```python
plt.imshow(bgrg)
plt.show()
```

```python
plt.imshow(bgrr)
plt.show()
```
